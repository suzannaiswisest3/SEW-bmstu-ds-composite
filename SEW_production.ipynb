````python
%%writefile README.md
# Predicting End Properties of Composite Materials

This repository contains an end-to-end machine learning and deep learning pipeline designed to predict the mechanical properties of composite materials and provide reverse-engineering recipe recommendations. The project is implemented as part of a research initiative to optimize manufacturing parameters.

## 📊 Project Structure & Tasks

The project solves two interconnected technological problems based on a dataset of physical, chemical, and mechanical parameters:

1. **Predicting Mechanical Properties (Regression Task):**
   * **Targets:** Tensile Modulus (GPa) and Tensile Strength (MPa).
   * **Best Model:** Optimized **Random Forest Regressor** with Hyperparameter Tuning via Cross-Validation ($R^2 > 0.14$).
   * **Baselines Evaluated:** Ridge Regression (Polynomial Features), XGBoost, and CatBoost.

2. **Recipe Recommendation System (Deep Learning Task):**
   * **Target:** Matrix-to-Filler Ratio.
   * **Model Architecture:** Multi-Layer Perceptron (MLP) built with TensorFlow/Keras.
   * **Topology:** Two hidden layers (8x8 neurons) with Batch Normalization and Dropout to prevent overfitting on noisy experimental data.
   * **Optimization:** Noise reduction via Multi-dimensional Anomaly Detection (Isolation Forest) and Principal Component Analysis (PCA) successfully lifted the model into positive $R^2$ performance territory.

## 📂 Repository Contents

* `matrix_filler_ratio_recommendation_nn.keras` - Fully trained Keras deep learning model for recipe recommendation.
* `best_random_forest_model.pkl` - Serialized ensemble model for predicting tensile strength and modulus.
* `scaler_nn_clean.pkl` & `pca_nn_clean.pkl` - Serialized preprocessing pipelines for data normalization and PCA dimensionality reduction.
* `feature_names.pkl` - Reference vector maintaining strict entry feature ordering.
* `requirements.txt` - Configuration file locking all library versions to guarantee full environment reproducibility.

## 🚀 Quick Start & Installation

To deploy this analytical package locally or inside an isolated server session, follow these steps:

1. **Clone the repository:**
   ```bash
   git clone github.com
   cd composite-properties-prediction
   ```

2. **Install all required dependencies with strict version matching:**
   ```bash
   pip install -r requirements.txt
   ```

3. **Verify model deployment in Python:**
   ```python
   import joblib
   from tensorflow.keras import models

   # --- LOAD NEURAL NETWORK COMPONENTS ---

   # Эта нейросеть рекомендует оптимальное значение для целевой переменной: 'Соотношение матрица-наполнитель'
   model = models.load_model('matrix_filler_ratio_recommendation_nn.keras')

   # Данный модуль нормализует входные параметры перед их подачей в рекомендательную нейросеть
   scaler = joblib.load('scaler_nn_clean.pkl')

   # Данный модуль сжимает 12 входных признаков через метод PCA для удаления шума перед работой нейросети
   pca = joblib.load('pca_nn_clean.pkl')

   # --- LOAD MACHINE LEARNING MODEL COMPONENTS ---

   # Эта модель Случайного леса синхронно предсказывает две целевые переменные:
   # 'Модуль упругости при растяжении, ГПа' и 'Прочность при растяжении, МПа'
   ml_model = joblib.load('best_random_forest_model.pkl')

   # Данный модуль нормализует входные технологические признаки для пайплайна Случайного леса
   ml_scaler = joblib.load('scaler.pkl')

   print("All analytics cores and predictive models loaded successfully!")
   ```

## 📈 Methodology Summary

* **Data Preprocessing:** Table merging using an `INNER JOIN` logic, custom soft threshold outlier removal, and automated multidimensional noise filtering via `Isolation Forest`.
* **Feature Engineering:** Scaled normalization combined with `PCA` retaining 90% of structural variance to eliminate high-frequency experiment noise.
* **Validation Strategy:** Rigorous 5-fold cross-validation (`KFold`) to eliminate sampling bias and guarantee generalization across different production batches.
````
